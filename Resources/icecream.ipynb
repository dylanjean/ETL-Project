```python
name = []
zipcode =[]
address = []
rating=[]
review = []
price = []
j=0

#URL for Yelp's API
url='https://api.yelp.com/v3/businesses/search'

#looping the API and collecting necessary data
  
param = {'location':"Philadelphia",'term':'icecream','limit':50}
responses=requests.get(url, params=param, headers=headers).json()


for bus in responses['businesses']:

    try:
        name.append(responses['businesses'][j]['name'])
        zipcode.append(responses['businesses'][j]['location']['zip_code'])
        rating.append(responses['businesses'][j]['rating'])
        review.append(responses['businesses'][j]['review_count'])
        price.append(responses['businesses'][j]['price'])
    except:
        price.append("")
    j+=1
            

    
```

```python
#creating and printing out the dataframe
icecream_df = pd.DataFrame({"Business Name":name,"Zip Code":zipcode,"Review Count":review,"Rating":rating, "Price":price})

icecream_df['Price Numerical'] =""
for index, row in icecream_df.iterrows():
    if icecream_df.loc[index,'Price'] == "$":
        icecream_df.loc[index,'Price Numerical'] = 1
    elif icecream_df.loc[index,'Price'] == "$$":
        icecream_df.loc[index,'Price Numerical'] = 2
    elif icecream_df.loc[index,'Price'] == "$$$":
        icecream_df.loc[index,'Price Numerical'] = 3
    elif icecream_df.loc[index,'Price'] == "$$$$":
        icecream_df.loc[index,'Price Numerical'] = 4


# icecream_df["Zip Code"]=icecream_df["Zip Code"].astype(int)
icecream_df
```
